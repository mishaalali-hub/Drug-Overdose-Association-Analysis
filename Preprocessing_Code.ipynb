```python
# 1: IMPORT LIBRARIES + BASIC SETTINGS
# FUNCTION:
#   Loads all libraries needed for preprocessing + frequency/co-occurrence analysis
# MECHANISM:
#   Imports pandas/numpy for data operations; matplotlib/seaborn for plots.
# EXPECTED OUTPUT:
#   A confirmation message without errors.
# CONSIDERATIONS:
#   If a module is missing, restart runtime and re-run.


import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from IPython.display import display

pd.set_option("display.max_columns", 200)
pd.set_option("display.width", 140)

print(" Libraries imported and display settings configured.")
```

```python
# 2: LOAD DATASET + SHOWING FIRST 20 ROWS
# FUNCTION:
#   Loads Accidental_Drug_Related_Deaths.csv and shows first 20 rows.
# MECHANISM:
#   Reads CSV into a DataFrame using pandas.
# EXPECTED OUTPUT:
#   Dataset shape + first 20 rows table.
# CONSIDERATIONS:
#   Update DATA_PATH if your file is elsewhere in Drive.

from google.colab import drive
drive.mount('/content/drive')

DATA_PATH = "/content/drive/MyDrive/Accidental_Drug_Related_Deaths.csv"

df_raw = pd.read_csv(DATA_PATH)

print(" Loaded:", DATA_PATH)
print(" Shape:", df_raw.shape)

# REQUIRED screenshot for report
display(df_raw.head(20))
```

```python
# 3: STANDARDIZE COLUMN NAMES (CONSISTENCY)
# FUNCTION:
#   Normalizes column names so later selection (drug_cols) never mismatches.
# MECHANISM:
#   Lowercase + strip + replace spaces/hyphens + remove parentheses.
# EXPECTED OUTPUT:
#   df columns become consistent snake_case.
# CONSIDERATIONS:
#   Always run this after loading df_raw to prevent KeyError.

df = df_raw.copy()

df.columns = (
    df.columns
      .str.strip()
      .str.lower()
      .str.replace(" ", "_", regex=False)
      .str.replace("-", "_", regex=False)
      .str.replace("/", "_", regex=False)
      .str.replace("(", "", regex=False)
      .str.replace(")", "", regex=False)
)

print(" Column names standardized.")
print(" Example columns:", df.columns[:10].tolist())
```

```python
# 4: REMOVE DUPLICATES
# FUNCTION:
#   Removes duplicate rows to avoid biased frequency/co-occurrence counts.
# MECHANISM:
#   drop_duplicates() removes exact duplicate records.
# EXPECTED OUTPUT:
#   Printed number of removed duplicates and updated shape.
# CONSIDERATIONS:
#   If duplicates represent legitimate repeated cases, discuss in report; commonly they
#   are data-entry duplicates, so removal is reasonable.

before = len(df)
df = df.drop_duplicates()
after = len(df)

print(f" Duplicates removed: {before - after}")
print(" Shape after duplicate removal:", df.shape)
```

```python
# 5: IDENTIFY DRUG INDICATOR COLUMNS
# FUNCTION:
#   Detects columns that represent drug presence indicators.
# MECHANISM:
#   Uses keyword matching against column names.
# EXPECTED OUTPUT:
#   Printed list of detected drug columns.
# CONSIDERATIONS:
#   You may manually adjust the list if some columns are incorrectly included/excluded.

drug_keywords = [
    "heroin", "fentanyl", "cocaine", "benzodia", "oxycodone", "methadone",
    "ethanol", "hydrocodone", "morphine", "amphetamine",
    "opioid", "opiate", "buprenorphine", "tramadol", "marijuana"
]

drug_cols_auto = []
for c in df.columns:
    if any(kw in c for kw in drug_keywords):
        drug_cols_auto.append(c)

drug_cols_auto = sorted(set(drug_cols_auto))

print(" Detected drug-related columns (auto):")
print(drug_cols_auto)

# Use auto list by default
drug_cols = drug_cols_auto
```

```python
# A1: DATA QUALITY SUMMARY
# FUNCTION:
#   Provides a clear snapshot of missing values and duplicates before/after cleaning.
# MECHANISM:
#   - isna().sum() identifies missing values per column
#   - duplicates counted from raw data
# EXPECTED OUTPUT:
#   Tables showing missingness (top columns) and duplicates removed.
# CONSIDERATIONS:
#   Taking screenshots of outputs for report.

# Missing values (Top 15)
missing = df.isna().sum().sort_values(ascending=False)
missing_table = missing[missing > 0].head(15).reset_index()
missing_table.columns = ["column", "missing_count"]

print(" Top columns with missing values (after non-drug imputation; drug NaNs preserved):")
display(missing_table)

# Duplicate evidence (from raw)
dup_count_raw = df_raw.duplicated().sum()
print(f" Duplicate rows in raw dataset: {dup_count_raw}")
print(f" Rows after removing duplicates: {len(df)}")
```

```python
# 6: HANDLE MISSING VALUES (NON-DRUG COLUMNS ONLY)
# FUNCTION:
#   Improves data quality without corrupting drug presence information.
# MECHANISM:
#   - Non-drug numeric refers to median
#   - Non-drug categorical would refer "Unknown"
# EXPECTED OUTPUT:
#   Reduced missingness in non-drug fields; drug NaNs preserved (meaning "not present").
# CONSIDERATIONS:
#   In this dataset, drug columns commonly use NaN to represent absence. Imputing those
#   would incorrectly mark substances as present.

non_drug_cols = df.columns.difference(drug_cols)

num_cols_non_drug = df[non_drug_cols].select_dtypes(include=[np.number]).columns.tolist()
cat_cols_non_drug = df[non_drug_cols].select_dtypes(exclude=[np.number]).columns.tolist()

for c in num_cols_non_drug:
    df[c] = df[c].fillna(df[c].median())

for c in cat_cols_non_drug:
    df[c] = df[c].fillna("Unknown")

print(" Missing values handled for NON-drug columns only.")
print(" Remaining missing values (mostly drug NaNs by design):", int(df.isna().sum().sum()))
```

```python
drug_cols_existing = [c for c in drug_cols if c in df.columns]

df_drugs = df[drug_cols_existing].copy()

for c in df_drugs.columns:
    df_drugs[c] = df_drugs[c].notna().astype(int)
```

```python
# A2: DRUG MATRIX VALIDATION
# FUNCTION:
#   Confirms df_drugs contains meaningful 0/1 patterns (NOT all ones).
# MECHANISM:
#   - mean() shows prevalence per drug
#   - row sums show how many substances per case
# EXPECTED OUTPUT:
#   Drug prevalence stats + distribution of substances per case.
# CONSIDERATIONS:
#   Screenshotting the results as they justify that preprocessing is correct.

# Prevalence per drug (percentage)
drug_prevalence = (df_drugs.mean() * 100).sort_values(ascending=False).round(2)
print(" Drug prevalence (%) - Top 15:")
display(drug_prevalence.head(15))

# Substances per case distribution
substances_per_case = df_drugs.sum(axis=1)

print(" Substances per case (summary):")
display(substances_per_case.describe())

print(" Distribution (counts) of substances per case (top 15 values):")
display(substances_per_case.value_counts().head(15))
```

```python
# A3: FINAL MINING-READY DATASET
# FUNCTION:
#   Builds a final dataset that is ready for analysis and submission.
# MECHANISM:
#   - Drop raw drug columns (text/NaN format)
#   - Append binary drug indicators
# EXPECTED OUTPUT:
#   clean_df with consistent features.
# CONSIDERATIONS:
#   This ensures me that the report can clearly state what was transformed and why.

# Remove original drug columns and replace with binary versions
clean_df = df.drop(columns=drug_cols_existing).copy()
clean_df = pd.concat([clean_df.reset_index(drop=True), df_drugs.reset_index(drop=True)], axis=1)

print(" clean_df created (non-drug cleaned + drug binary indicators).")
print(" clean_df shape:", clean_df.shape)
display(clean_df.head())
```

```python
# 7: BUILD BINARY DRUG MATRIX (0/1)
# FUNCTION:
#   Converts drug indicator columns into a consistent binary matrix.
# MECHANISM:
#   For this dataset: NaN = absent is 0, non-null = present is 1.
# EXPECTED OUTPUT:
#   df_drugs with 0/1 values; NOT all ones.
# CONSIDERATIONS:
#   If a drug column is missing from df for any reason, we filter it safely.

drug_cols_existing = [c for c in drug_cols if c in df.columns]
print(" Drug columns found:", len(drug_cols_existing))

df_drugs = df[drug_cols_existing].copy()

for c in df_drugs.columns:
    df_drugs[c] = df_drugs[c].notna().astype(int)

print(" Drug indicator columns standardized to binary 0/1.")
display(df_drugs.head())

# Sanity check: should NOT show 100% for everything
print("\n Top 15 substances by frequency (sanity check):")
display(df_drugs.sum().sort_values(ascending=False).head(15))

plt.figure(figsize=(8,5))
substances_per_case.plot(kind='hist', bins=10)
plt.title("Distribution of Number of Substances per Case")
plt.xlabel("Number of substances")
plt.ylabel("Frequency")
plt.show()
```

```python
# 8: SUBSTANCE FREQUENCY ANALYSIS
# FUNCTION:
#   Computes and visualizes how often each substance appears in cases.
# MECHANISM:
#   Column sums of df_drugs provide case counts which are divided by total cases for %.
# EXPECTED OUTPUT:
#   Frequency table
# CONSIDERATIONS:
#   Percentages are calculated using number of cases after cleaning/duplicate removal.

n_cases = len(df_drugs)

drug_freq = df_drugs.sum(axis=0).sort_values(ascending=False)
drug_pct = (drug_freq / n_cases * 100).round(2)

freq_table = pd.DataFrame({
    "substance": drug_freq.index,
    "case_count": drug_freq.values,
    "percentage_of_cases": drug_pct.values
})

print(" Substance frequency table (top 20):")
display(freq_table.head(20))
```

```python
# B1: VISUALIZATION — TOP SUBSTANCES
# FUNCTION:
#   Visualizes the most frequently involved substances.
# MECHANISM:
#   Column sums represent number of cases where substance is present.
# EXPECTED OUTPUT:
#   Bar chart with top substances.
# CONSIDERATIONS:
#   Used with interpretation (not just description).

drug_counts = df_drugs.sum().sort_values(ascending=False)
top_k = 15

plt.figure(figsize=(10, 6))
sns.barplot(x=drug_counts.head(top_k).values, y=drug_counts.head(top_k).index)
plt.title("Top 15 Substances by Case Count")
plt.xlabel("Number of cases where substance is present")
plt.ylabel("Substance")
plt.show()
```

```python
# B2: VISUALIZATION — POLY-DRUG INVOLVEMENT
# FUNCTION:
#   Shows how common multi-substance involvement is.
# MECHANISM:
#   Row sums = number of substances present per case.
# EXPECTED OUTPUT:
#   Bar chart of single vs multi vs none.
# CONSIDERATIONS:
#   Strong interpretation value for co-occurrence discussion.

substances_per_case = df_drugs.sum(axis=1)

counts = pd.Series({
    "No flagged substance": (substances_per_case == 0).sum(),
    "Single substance": (substances_per_case == 1).sum(),
    "Poly-drug (2+)": (substances_per_case >= 2).sum()
})

plt.figure(figsize=(8, 5))
sns.barplot(x=counts.index, y=counts.values)
plt.title("Distribution of Cases by Number of Substances")
plt.xlabel("Case type")
plt.ylabel("Number of cases")
plt.xticks(rotation=15)
plt.show()

display(counts.to_frame("case_count"))
```

```python
# 9: CO-OCCURRENCE ANALYSIS (PAIR LEVEL)
# FUNCTION:
#   Identifies poly-drug cases and the most common co-occurring substance pairs.
# MECHANISM:
#   - Row sum -> number of substances per case
#   - X.T @ X -> co-occurrence matrix
# EXPECTED OUTPUT:
#   Counts of poly-drug cases + top pairs table + bar chart.
# CONSIDERATIONS:
#   Pair co-occurrence is symmetric so we keep upper triangle only.

substances_per_case = df_drugs.sum(axis=1)

multi_drug_cases = int((substances_per_case >= 2).sum())
single_drug_cases = int((substances_per_case == 1).sum())
zero_drug_cases = int((substances_per_case == 0).sum())

print(" Case-level substance counts:")
print(" - Poly-drug cases (>=2 substances):", multi_drug_cases)
print(" - Single-substance cases:", single_drug_cases)
print(" - No flagged substances:", zero_drug_cases)

# Co-occurrence matrix
co_matrix = df_drugs.T.dot(df_drugs)

pairs = []
cols = co_matrix.columns.tolist()

for i in range(len(cols)):
    for j in range(i + 1, len(cols)):
        pairs.append((cols[i], cols[j], int(co_matrix.iloc[i, j])))

pairs_df = pd.DataFrame(pairs, columns=["item_1", "item_2", "cooccurrence_count"])
pairs_df = pairs_df.sort_values("cooccurrence_count", ascending=False)

print("\n Top 20 co-occurring substance pairs:")
display(pairs_df.head(20))

# Plot top pairs
top_pairs = pairs_df.head(15).copy()
top_pairs["pair"] = top_pairs["item_1"] + " + " + top_pairs["item_2"]

plt.figure(figsize=(10, 6))
sns.barplot(x=top_pairs["cooccurrence_count"], y=top_pairs["pair"])
plt.title("Top Co-occurring Substance Pairs")
plt.xlabel("Number of cases where both substances are present")
plt.ylabel("Substance Pair")
plt.show()
```

```python
# B3: CO-OCCURRENCE HEATMAP
# FUNCTION:
#   Visualizes pairwise co-occurrence between top substances.
# MECHANISM:
#   Use top N substances; plot their co-occurrence matrix heatmap.
# EXPECTED OUTPUT:
#   Heatmap
# CONSIDERATIONS:
#   Used for stronger visuals

top_n = 10
top_substances = drug_counts.head(top_n).index.tolist()

co_top = co_matrix.loc[top_substances, top_substances]

plt.figure(figsize=(8, 6))
sns.heatmap(co_top, annot=True, fmt="d")
plt.title(f"Co-occurrence Heatmap (Top {top_n} Substances)")
plt.xlabel("Substance")
plt.ylabel("Substance")
plt.show()
```

```python
# CELL 10: CREATE TRANSACTIONS FOR APRIORI / FP-GROWTH
# FUNCTION:
#   Converts each case into a list of substances present (transaction format).
# MECHANISM:
#   For each row, collect column names where value == 1.
# EXPECTED OUTPUT:
#   df_transactions with items + item_count, and a non-empty subset.
# CONSIDERATIONS:
#   Empty transactions (item_count=0) are typically removed before Apriori.

transactions = []
for _, row in df_drugs.iterrows():
    items = row.index[row == 1].tolist()
    transactions.append(items)

df_transactions = pd.DataFrame({
    "items": [",".join(items) for items in transactions],
    "item_count": [len(items) for items in transactions]
})

print(" Transaction dataset created.")
display(df_transactions.head(10))

df_transactions_nonempty = df_transactions[df_transactions["item_count"] > 0].reset_index(drop=True)
print(" Non-empty transactions:", len(df_transactions_nonempty), "out of", len(df_transactions))
```

```python
# CELL 11: SAVE OUTPUT FILES
# FUNCTION:
#   Saved:
#   - cleaned dataset (Excel) for submission
#   - drug binary matrix (CSV) for evidence
#   - transactions (CSV) for Assignment 2 Apriori/FP-Growth
# MECHANISM:
#   Uses pandas export functions.
# EXPECTED OUTPUT:
#   Files created in Colab working directory.

df.to_excel("cleaned_dataset.xlsx", index=False)
df_drugs.to_csv("drug_binary_matrix.csv", index=False)
df_transactions.to_csv("transactions_all_cases.csv", index=False)
df_transactions_nonempty.to_csv("transactions_nonempty.csv", index=False)

print(" Saved files:")
print(" - cleaned_dataset.xlsx")
print(" - drug_binary_matrix.csv")
print(" - transactions_all_cases.csv")
print(" - transactions_nonempty.csv")
```
